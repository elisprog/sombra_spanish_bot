```python
corpus_raw = []
for fname in ["1primero.txt", "2segundo.txt", "3tercero.txt", "4cuarto.txt", "5quinto.txt", "6sexto.txt", "7septimo.txt", "8octavo.txt", "9noveno.txt", "10decimo.txt", "11undecimo.txt", "12duodecimo.txt", "13decimotercero.txt"]:
    with open(fname, encoding="utf-8") as f:
        corpus_raw.append(f.read())

preprocessed_texts = [preprocess(text) for text in corpus_raw]
```

```python
from gensim import corpora
from gensim.models.ldamodel import LdaModel
import re

dictionary = corpora.Dictionary(preprocessed_texts)
corpus = [dictionary.doc2bow(text) for text in preprocessed_texts]

lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=5, passes=15, random_state=42)

words = set()
for topic in lda_model.print_topics(num_words=30):
    # topic[1] содержит строку вида: '0.013*"слово1" + 0.011*"слово2" + ...'
    words.update(re.findall(r'"(.*?)"', topic[1]))

print(f"Количество уникальных слов: {len(words)}")
print(words)
```

```python
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

lda_display = gensimvis.prepare(lda_model, corpus, dictionary)

# Отображение в Jupyter Notebook
pyLDAvis.enable_notebook()
pyLDAvis.display(lda_display)
```
